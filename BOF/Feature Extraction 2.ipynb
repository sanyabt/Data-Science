```python
import pandas as pd
```

```python
train_data=pd.read_csv('/Users/Seema/Desktop/ML Workshop/BOF/labeledTrainData.tsv', sep='\t')
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vocab = ['awesome','good','nice','amazing','wonderful','like','interesting','terrible','hate','boring','awful']
```

```python
simple_vectorizer = CountVectorizer(vocabulary=vocab)
bow=simple_vectorizer.fit_transform(train_data.review).todense()
```

```python
bow
```

```python
words = list(simple_vectorizer.vocabulary_.keys())
```

```python
words
```

```python
bow_df = pd.DataFrame(bow, index=train_data.index, columns=words)
```

```python
bow_df[(bow_df.good==0)&(bow_df.like==0)&(bow_df.interesting==0)&(bow_df.awesome==0)&(bow_df.terrible==0)&(bow_df.amazing==0)&(bow_df.wonderful==0)&(bow_df.boring==0)&(bow_df.awful==0)&(bow_df.hate==0)&(bow_df.nice==0)].count()
```
