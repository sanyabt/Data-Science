```python
test_data = pd.read_csv('/Users/Seema/Desktop/ML Workshop/BOF/testData.tsv', sep='\t')
```

```python
test_data
```

```python
test_data['review_length']=test_data.review.apply(len)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vocab= ['awesome', 'good', 'nice', 'amazing', 'wonderful', 'like', 'interesting', 'terrible', 'hate', 'boring', 'awful']
n_vectorizer=CountVectorizer(vocabulary=vocab)
bow=n_vectorizer.fit_transform(test_data.review).todense()
words = list(n_vectorizer.vocabulary_.keys())
bow_df = pd.DataFrame(bow, index=test_data.index, columns= words)
test_data_expand=test_data.join(bow_df)
test_data_expand.to_csv('testDataFeatures.tsv', sep='\t')
```

```python
test_data_expand
```

```python
predictions = nb.predict(np.asarray(test_data_expand[columns]))
```

```python
submission = pd.DataFrame({"id": test_data.id, "sentiment": predictions})
```
